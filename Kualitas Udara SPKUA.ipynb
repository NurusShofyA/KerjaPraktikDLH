```python
baku_mutu = {
    "PM10": [(0, 50, 50), (50, 150, 100), (150, 350, 200), (350, 420, 300), (420, 500, 400)],
    "PM2.5": [(0, 15.5, 50), (15.5, 55.4, 100), (55.4, 150.4, 200), (150.4, 250.4, 300), (250.4, float('inf'), 400)],
    "SO2": [(0, 52, 50), (52, 180, 100), (180, 400, 200), (400, 800, 300), (800, 1200, 400)],
    "CO": [(0, 4000, 50), (4000, 8000, 100), (8000, 15000, 200), (15000, 30000, 300), (30000, 45000, 400)],
    "O3": [(0, 120, 50), (120, 235, 100), (235, 400, 200), (400, 800, 300), (800, 1000, 400)],
    "NO2": [(0, 80, 50), (80, 200, 100), (200, 400, 200), (400, 800, 300), (800, 1000, 400)],
    "HC": [(0, 45, 50), (45, 100, 100), (100, 215, 200), (215, 432, 300), (432, 648, 400)]
}

def calculate_ispu(value, pollutant):
    for lower_bound, upper_bound, ispu_value in baku_mutu[pollutant]:
        if lower_bound <= value < upper_bound:
            xx = value
            ia = ispu_value
            ib = ispu_value - 50  # ISPU untuk batas bawah
            xa = upper_bound
            xb = lower_bound

            ispu = (ia - ib) / (xa - xb) * (xx - xb) + ib
            return round(ispu, 1)
    return None
```

```python
Parameters['ISPU CO'] = Parameters['CO'].apply(lambda x: calculate_ispu(x, 'CO'))
```

```python
Parameters['ISPU PM10'] = Parameters['PM10'].apply(lambda x: calculate_ispu(x, 'PM10'))
Parameters['ISPU PM2.5'] = Parameters['PM2.5'].apply(lambda x: calculate_ispu(x, 'PM2.5'))
Parameters['ISPU SO2'] = Parameters['SO2'].apply(lambda x: calculate_ispu(x, 'SO2'))
Parameters['ISPU O3'] = Parameters['O3'].apply(lambda x: calculate_ispu(x, 'O3'))
Parameters['ISPU NO2'] = Parameters['NO2'].apply(lambda x: calculate_ispu(x, 'NO2'))
Parameters['ISPU HC'] = Parameters['HC'].apply(lambda x: calculate_ispu(x, 'HC'))
Parameters['Max ISPU']=Parameters[['ISPU PM10', 'ISPU PM2.5', 'ISPU SO2', 'ISPU CO', 'ISPU O3', 'ISPU NO2','ISPU HC']].max(axis=1)
Parameters
```

```python
plt.figure(figsize=(20,6))
plt.plot(Parameters['Tanggal'], Parameters['Max ISPU'],marker='o',  color='#013D5A')
plt.title('Trend Indeks Standar Pencemaran Udara Selama 2024')
plt.xlabel('Date')
plt.ylabel('ISPU')
ticks = Parameters['Tanggal'][::10]
plt.xticks(ticks, rotation=45, fontsize=10)
plt.axhline(y=50, color='g', linestyle='--', label='Baik')
plt.axhline(y=100, color='b', linestyle='--', label='Sedang')
plt.axhline(y=199, color='y', linestyle='--', label='Tidak Sehat')
plt.axhline(y=299, color='r', linestyle='--', label='Sangat Tidak Sehat')
plt.fill_between(Parameters['Tanggal'], 0, 50, color='g', alpha=0.3)  
plt.fill_between(Parameters['Tanggal'], 51, 100, color='b', alpha=0.3)  
plt.fill_between(Parameters['Tanggal'], 101, 199, color='y', alpha=0.3)  
plt.fill_between(Parameters['Tanggal'], 200, 299, color='r', alpha=0.3) 
plt.fill_between(Parameters['Tanggal'], 300, 400, color='black', alpha=0.3)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.25), ncol=4, title="Kategori ISPU", frameon=False)
plt.show()
```
